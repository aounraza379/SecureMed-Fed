```python
import pandas as pd
import numpy as np
import random

# 1. Set the number of data points (simulating about 15-16 minutes of data)
data_points = 1000

# 2. Generate "Normal" heart rates (60 to 90 beats per minute)
heart_rates = [random.randint(60, 90) for _ in range(data_points)]

# 3. Create a "Cardiac Anomaly" (Emergency)
# We will change the last 50 readings to be dangerously high
for i in range(950, 1000):
    heart_rates[i] = random.randint(150, 180)

# 4. Create a Timestamp for each reading
timestamps = pd.date_range(start="2026-04-01 08:00:00", periods=data_points, freq='S')

# 5. Combine into a Spreadsheet (DataFrame)
df = pd.DataFrame({
    'Timestamp': timestamps,
    'Heart_Rate': heart_rates
})

# 6. Save it as a CSV file to use later
df.to_csv('patient_data.csv', index=False)

print("Phase 1 Complete: 'patient_data.csv' has been created.")
print(df.tail(10)) # Shows the high heart rates we created
```
